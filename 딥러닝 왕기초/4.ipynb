```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for plotting
x = np.linspace(-10, 10, 500)

# Unit step function
unit_step = np.heaviside(x, 1)

# Sigmoid function
sigmoid = 1 / (1 + np.exp(-x))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, unit_step, label="Unit Step Function", linestyle='--', color='blue')
plt.plot(x, sigmoid, label="Sigmoid Function", color='red')
plt.title("Unit Step Function and Sigmoid Function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid for 3D plot
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Unit step function
Z_unit_step = np.where(X + Y > 5, 1, 0)  # Example step condition: X + Y > 5

# Sigmoid function
Z_sigmoid = 1 / (1 + np.exp(-0.5 * (X + Y - 5)))  # Shifted for comparison

# Plot the functions
fig = plt.figure(figsize=(12, 6))

# Unit step function plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_unit_step, cmap='viridis', alpha=0.8)
ax1.set_title('Unit Step Function (3D)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Sigmoid function plot
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_sigmoid, cmap='plasma', alpha=0.8)
ax2.set_title('Sigmoid Function (3D)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
w = np.linspace(-5, 5, 500)  # Narrower range for w
sigmoid = 1 / (1 + np.exp(-w))
mse_loss = (sigmoid - 1)**2
bce_loss = -np.log(sigmoid)

# Plot the graphs
plt.figure(figsize=(8, 6))
plt.plot(w, mse_loss, label=r'MSE Loss: $(\frac{1}{1 + e^{-w}} - 1)^2$', linewidth=2)
plt.plot(w, bce_loss, label=r'BCE Loss: $-\log(\frac{1}{1 + e^{-w}})$', linewidth=2, linestyle='--')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.ylim(0, 2)  # Adjusting the y-axis for better visualization
plt.xlabel(r'$w$', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.title('MSE Loss vs BCE Loss (narrowed $w$ range)', fontsize=16)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sigmoid function
def sigmoid(w):
    return 1 / (1 + np.exp(-w))

# Define second derivative of MSE Loss
def mse_second_derivative(w):
    s = sigmoid(w)
    return 2 * s * (1 - s) * (1 - 6 * s + 6 * s**2)

# Define second derivative of BCE Loss
def bce_second_derivative(w):
    s = sigmoid(w)
    return s * (1 - s)

# Generate values for w
w = np.linspace(-5, 5, 500)

# Calculate second derivatives
mse_second = mse_second_derivative(w)
bce_second = bce_second_derivative(w)

# Plot second derivatives
plt.figure(figsize=(10, 6))
plt.plot(w, mse_second, label="MSE Loss (2nd Derivative)", color="blue", linestyle="--")
plt.plot(w, bce_second, label="BCE Loss (2nd Derivative)", color="red")
plt.axhline(0, color="black", linestyle=":", linewidth=1)
plt.title("Second Derivatives of MSE and BCE Loss", fontsize=14)
plt.xlabel("w", fontsize=12)
plt.ylabel("Second Derivative", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
