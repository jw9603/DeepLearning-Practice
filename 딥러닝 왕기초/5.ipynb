```python
# Re-import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the Unit Step Function
def unit_step(x, threshold):
    return np.where(x > threshold, 1, 0)

# Generate x values for visualization
x = np.linspace(0, 2000, 500)

# Define thresholds for the two neurons
thresholds = [900, 1100]

# Compute the unit step outputs
y1 = 1000 * unit_step(x, thresholds[0])
y2 = -1000 * unit_step(x, thresholds[1])

# Plot the unit step functions
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label=f'Neuron 1 (Threshold={thresholds[0]})', color='blue', linewidth=2)
plt.plot(x, y2, label=f'Neuron 2 (Threshold={thresholds[1]})', color='green', linewidth=2)

plt.title('Unit Step Function for Two Neurons', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.axvline(x=thresholds[0], color='blue', linestyle='--', alpha=0.6)
plt.axvline(x=thresholds[1], color='green', linestyle='--', alpha=0.6)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
# Plotting the sum of two Unit Step Functions

# Sum of the two unit step functions
y_combined = y1 + y2

# Plot the individual functions and their sum
plt.figure(figsize=(8, 6))

# plt.step(x, y1, label="Neuron 1 (Threshold = 900)", where='post')
# plt.step(x, y2, label="Neuron 2 (Threshold = 1100)", where='post')
plt.step(x, y_combined, label="Combined Output", where='post', linestyle='--', color='purple')

plt.title("Unit Step Functions and Their Combined Output")
plt.xlabel("Input Value (x)")
plt.ylabel("Output Value")
plt.axhline(1, color="gray", linestyle="--", linewidth=0.5)
plt.axhline(2, color="gray", linestyle="--", linewidth=0.5)
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
