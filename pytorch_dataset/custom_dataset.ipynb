### Convert to PyTorch Tensor

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data = torch.from_numpy(df.values).float()
data.shape
```

```python
x = data[:,:-1]
y = data[:,-1:]

print(f'x:{x.shape}')
print(f"y:{y.shape}")
```

```python
ratios = [0.6,0.2,0.2]
```

```python
train_cnt = int(data.size(0) * ratios[0])
valid_cnt = int(data.size(0) * ratios[1])
test_cnt = data.size(0) - train_cnt - valid_cnt

print(f"Train : {train_cnt}, Valid : {valid_cnt}, Test : {test_cnt} samples")
```

```python
indices = torch.randperm(data.size(0))
indices
```

```python
cnts = [train_cnt, valid_cnt, test_cnt]
```

```python
x = torch.index_select(x, dim=0, index=indices)
y = torch.index_select(y, dim=0, index=indices)

x = x.split(cnts, dim=0)
y = y.split(cnts, dim=0)

for x_i, y_i in zip(x, y):
    print(x_i.shape, y_i.shape)
```

### Set HyperParameters

```python
n_epochs = 10000
batch_size = 64
print_interval = 200
early_stop = 100
layers = 6
```

### Get DataLoaders

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CancerDataset(Dataset):
    
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        
        super().__init__()
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self,idx):
        return self.data[idx], self.labels[idx]
```

```python
train_loader = DataLoader(
    dataset=CancerDataset(x[0],y[0]),
    batch_size=batch_size,
    shuffle=True) # Train data는 셔플을 진행해야 함.

valid_loader = DataLoader(
    dataset=CancerDataset(x[1],y[1]),
    batch_size=batch_size,
    shuffle=False)

test_loader = DataLoader(
    dataset=CancerDataset(x[2],y[2]),
    batch_size=batch_size,
    shuffle=False)

print(f"Train: {len(train_loader.dataset)}, Valid: {len(valid_loader.dataset)}, Test: {len(test_loader.dataset)}")
```

### Build Model and Optimizer

```python
layer_sizes = [x[0].size(-1) - i * (x[0].size(-1) - y[0].size(-1)) // (layers - 1) for i in range(layers)]
print(layer_sizes)
```

```python
model_layers = []
for i in range(len(layer_sizes)-1):
    model_layers.append(nn.Linear(layer_sizes[i],layer_sizes[i+1]))
    model_layers.append(nn.LeakyReLU())

model_layers.append(nn.Sigmoid())

model = nn.Sequential(*model_layers)
```

```python
model
```

```python
optimizer = optim.Adam(model.parameters())
```

### Train

```python
from copy import deepcopy

lowest_loss = np.inf
best_model = None
lowest_epoch = np.inf # it is for early stopping
```

```python
train_history, valid_history = [], []

for i in range(n_epochs):
    model.train()
    
    train_loss = 0
    valid_loss = 0
    
    for x_i, y_i in train_loader:
        # |x_i| = (batch size, 10)
        y_hat_i = model(x_i) 
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        optimizer.zero_grad()
        loss.backward()
        
        optimizer.step()
        train_loss += float(loss)
    
    train_loss = train_loss / len(train_loader)
    
    
    # 여기서부터 검증 데이터셋으로 검증
    model.eval()
    with torch.no_grad():
        
        for x_i, y_i in valid_loader:
            y_hat_i = model(x_i)
            loss = F.binary_cross_entropy(y_hat_i, y_i)
            
            valid_loss += float(loss)
            
    valid_loss = valid_loss / len(valid_loader)
    
    train_history.append(train_loss)
    valid_history.append(valid_loss)
    
    if (i+1) % print_interval == 0:
        print(f"Epoch: {i+1}: train loss: {train_loss:.4e}, valid loss: {valid_loss:.4e}, lowest valid loss: {lowest_loss:.4e}")
        
    
    if valid_loss <= lowest_loss:
        lowest_loss = valid_loss
        lowest_epoch = i
        
        best_model = deepcopy(model.state_dict())
    
    else:
        if early_stop > 0 and lowest_epoch + early_stop < i + 1:
            print(f"There is no improvement during last {early_stop} epochs.")
            break

print(f"Best validation loss from epoch: {lowest_epoch + 1}, {lowest_loss}")

model.load_state_dict(best_model)
```

### Loss Figure

```python
plt.figure(figsize=(30,15))
plt.grid(True)
plt.title(f"Train / Valid Loss History")

plt.plot(
    range(1, len(train_history)),train_history[1:],
    range(1, len(valid_history)),valid_history[1:],

)
plt.legend('True',fontsize=30)
plt.yscale('log')
plt.show()
```

### Test Result

```python
test_loss = 0
y_hat = []

model.eval()
with torch.no_grad():
    for x_i, y_i in test_loader:
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        test_loss += loss
        
        y_hat.append(y_hat_i)

test_loss = test_loss / len(test_loader)

y_hat = torch.cat(y_hat,dim=0)

print(y_hat.shape)

print(f"Test loss: {test_loss:.4e}")
```

```python
correct_cnt = (y[2] == (y_hat > .5)).sum()

total_cnt = y[2].size(0)

print(f"Test Acc: {correct_cnt / total_cnt * 100:2f} %")
```
